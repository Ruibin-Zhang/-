```python
damage_sample_s = convert_to_SimpleIndex(damage_sample, axis=1)
damage_sample_s
```

```python
if 'collapse-1' in damage_sample_s.columns:
    damage_sample_s['collapses/collapsed'] = damage_sample_s['collapse-1']
else:
    damage_sample_s['collapses/collapsed'] = np.zeros(damage_sample_s.shape[0])

if 'excessiveRID-1' in damage_sample_s.columns:
    damage_sample_s['reconstruction/irreparable'] = damage_sample_s['excessiveRID-1']
else:
    damage_sample_s['reconstruction/irreparable'] = np.zeros(damage_sample_s.shape[0])

agg_repair_s = convert_to_SimpleIndex(agg_repair, axis=1)

summary = pd.concat([agg_repair_s,
                        damage_sample_s[['collapses/collapsed', 'reconstruction/irreparable']]],
                    axis=1)
```

```python
summary["collapses/collapsed"] = np.where(summary["collapses/collapsed"] != 0, 1, 0)
summary["reconstruction/irreparable"] = np.where(
    (summary["reconstruction/irreparable"] != 0) & (summary["collapses/collapsed"] == 0),
    1,
    0
)
summary["reconstruction/cost_impractical"] = np.where(summary["repair_cost-"] > total_cost, 1, 0)
# summary["reconstruction/cost_impractical"] = 0
summary
```

```python
# save summary sample
summary.to_csv("DL_summary.csv", index_label='#Num')
```
